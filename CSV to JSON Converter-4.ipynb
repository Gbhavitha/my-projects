```python
#CSV to JSON converter
```

```python
#importing csv files and json module
import csv
import json
f1 = open("sales_data.csv","r")   #opens the file
read = csv.reader(f1)

csv_data = [row for row in read]
for line in csv_data:
    print(line)
    f1.close()  # Close the file
```

```python
f2 = open("sales_data1.json","w")  #creates the file 'sales_data1.json' in write mode 
json.dump(csv_data,f2)   
f2.close() 
f3 = open("sales_data1.json","r")   # Opens the file 'sales_data1.json' in read mode 
data = json.load(f3)  
for i in data:
    print(i)  # Prints each item
```
